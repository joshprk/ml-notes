# ZCA Whitening

Assume we found $W_{PCA}$, the PCA whitening matrix for a feature set $X$.

```python
import numpy as np
import matplotlib.pyplot as plt

dataset = []

for _ in range(500):
    dataset.append(np.random.normal(0, 1, 2))

dataset = np.array(dataset)
cov = np.cov(dataset.T)
eigenvalues, eigenvectors = np.linalg.eigh(cov)

pca_matrix = np.diag(1 / np.sqrt(eigenvalues)).dot(eigenvectors.T)
whitened_dataset = pca_matrix.dot(dataset.T)

plt.scatter(*whitened_dataset)
plt.show()
```

Any orthogonal rotation matrix $R$ such that $W = RW_{PCA}$ will produce a whitening matrix $W$.

Let $E$ be the diagonal eigenvalue matrix found in the EV decomposition $\Sigma = EDE^T$. Then, let $R = E$.

$W_{ZCA} = EW_{PCA}$

```python
zca_matrix = eigenvectors.dot(pca_matrix)
zca_whitened_dataset = zca_matrix.dot(dataset.T)

plt.scatter(*zca_whitened_dataset)
plt.show()
```

Note that the ZCA whitened dataset more closely resembles the original dataset than the PCA whitened dataset.
